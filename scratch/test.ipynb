```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.linspace(0,10,20)

b=[1 for i in range(20)]

plt.scatter(a,b,marker='.',color='r')
plt.show()
```

```python


b=[1+0.1*np.random.random() for i in range(20)]

plt.plot(a,b)
plt.show()
```

```python
def recursive_scatter(depth,start,end):
    if depth==0:
        return [end]
    else:
        mid=(start+end)/2+0.1*2**(depth-4)*np.random.random()
        return recursive_scatter(depth-1,start,mid)+recursive_scatter(depth-1,mid,end)
```

```python
a=recursive_scatter(4,1,1)
b=np.linspace(0,10,len(a))

plt.plot(b,a)
```
